Detectar y tratar valores atÃ­picos

```python
from scipy import stats
import numpy as np

# 1. TransformaciÃ³n Box-Cox mejorada
volume_cols = ['Total Volume', '4046', '4225', '4770', 'Total Bags', 
               'Small Bags', 'Large Bags', 'XLarge Bags']

# Crear copia para no modificar el original
avocado_clean = avocado.copy()

for col in volume_cols:
    # Box-Cox con ajuste de cero valores (si existen)
    data = avocado_clean[col] + 1  # Aseguramos > 0
    transformed, fitted_lambda = stats.boxcox(data)
    
    # Guardar la columna transformada
    avocado_clean[f'{col}_bc'] = transformed
    
    # Opcional: Inverse Box-Cox para verificaciÃ³n
    # original_approx = stats.inv_boxcox(transformed, fitted_lambda)
    print(f"{col}: Î» = {fitted_lambda:.3f}")
```

```python
# 3. DetecciÃ³n de outliers con IQR en datos transformados
bc_cols = [f'{col}_bc' for col in volume_cols] + ['AveragePrice']

# Usaremos 2.2*IQR como punto medio entre 1.5 (estricto) y 3 (conservador)
factor_iqr = 1.5

Q1 = avocado_clean[bc_cols].quantile(0.25)
Q3 = avocado_clean[bc_cols].quantile(0.75)
IQR = Q3 - Q1

filtro_outliers = ~((avocado_clean[bc_cols] < (Q1 - factor_iqr * IQR)) | 
                 (avocado_clean[bc_cols] > (Q3 + factor_iqr * IQR))).any(axis=1)

avocado_final = avocado_clean[filtro_outliers].copy()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 6))

# Antes
plt.subplot(1, 2, 1)
avocado_clean[volume_cols[:3]].boxplot()
plt.title('DistribuciÃ³n Original')

# DespuÃ©s
plt.subplot(1, 2, 2)
avocado_clean[[f'{col}_bc' for col in volume_cols[:3]]].boxplot()
plt.title('DistribuciÃ³n Post Box-Cox')

plt.tight_layout()
plt.show()

# Porcentaje de datos eliminados
print(f"\nDatos originales: {len(avocado_clean)}")
print(f"Datos limpios: {len(avocado_final)}")
print(f"Outliers eliminados: {100*(1 - len(avocado_final)/len(avocado_clean)):.2f}%")
```

```python
# Calcular porcentaje de outliers por variable
outliers_per_col = ((avocado_clean[bc_cols] < (Q1 - factor_iqr * IQR))) | \
                   (avocado_clean[bc_cols] > (Q3 + factor_iqr * IQR))

print("\nPorcentaje de outliers por columna:")
print((outliers_per_col.sum()/len(avocado_clean)*100))
```

```python
# FunciÃ³n para mostrar resumen estadÃ­stico comparativo
def compare_stats(original, cleaned, cols):
    print(f"{'EstadÃ­stica':<15}{'Original':>15}{'Limpio':>15}{'Diferencia':>15}")
    for col in cols:
        orig_mean = original[col].mean()
        clean_mean = cleaned[col].mean()
        print(f"{col:<15}"
              f"{orig_mean:>15.2f}"
              f"{clean_mean:>15.2f}"
              f"{(clean_mean-orig_mean):>15.2f}")

print("\nComparaciÃ³n de estadÃ­sticas (antes/despuÃ©s de eliminar outliers):")
compare_stats(avocado_clean, avocado_final, bc_cols)
```

ConversiÃ³n a Variables Dummy

```python
# Verificar variables categÃ³ricas
print("\n[1] AnÃ¡lisis de variables categÃ³ricas:")
print(f"- 'type': {avocado['type'].nunique()} categorÃ­as â†’ {avocado['type'].unique()}")
print(f"- 'region': {avocado['region'].nunique()} categorÃ­as")

# AnÃ¡lisis detallado de distribuciones
print("\nDistribuciÃ³n de categorÃ­as:")
print(avocado['type'].value_counts(normalize=True).apply(lambda x: f"{x:.1%}"))

region_counts = avocado['region'].value_counts()
print("\nTop 10 regiones mÃ¡s frecuentes:")
print(region_counts.head(10))
print("\n10 regiones menos frecuentes:")
print(region_counts.tail(10))
```

```python
# Hacemos una copia
avocado_optimized = avocado.copy()

# 1. Convertir 'type' igual que antes
avocado_optimized = pd.get_dummies(avocado_optimized, columns=['type'], prefix='type', drop_first=True)

# 2. Para 'region', agrupamos las menos frecuentes
region_counts = avocado_optimized['region'].value_counts()

# Definimos un umbral (ej: regiones con menos de 100 observaciones se agrupan como 'Other')
threshold = 100
less_frequent = region_counts[region_counts < threshold].index

# Creamos una columna agrupada
avocado_optimized['region_grouped'] = avocado_optimized['region'].replace(less_frequent, 'Other')

# Convertimos a dummies
avocado_optimized = pd.get_dummies(avocado_optimized, columns=['region_grouped'], prefix='region')

# Eliminamos la columna original de region
avocado_optimized.drop('region', axis=1, inplace=True)

# Verificamos
print("\nRegiones convertidas (agrupadas):")
print(avocado_optimized.filter(regex='region_').columns)
print("\nTotal de columnas:", len(avocado_optimized.columns))
```

```python
# Verificar muestras aleatorias
print("\nMuestra de datos convertidos:")
print(avocado_optimized[['type_organic'] + 
      list(avocado_optimized.filter(regex='region_').columns)].sample(5))

# Verificar proporciones
print("\nProporciÃ³n de tipos:")
print(avocado_optimized['type_organic'].value_counts(normalize=True))

print("\nRegiones mÃ¡s comunes:")
print(avocado_optimized.filter(regex='region_').sum().sort_values(ascending=False).head(10))
```
